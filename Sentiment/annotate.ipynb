```python
from transformers import pipeline
import json

# Model for sentiment analysis
distilled_student_sentiment_classifier = pipeline(
    model="lxyuan/distilbert-base-multilingual-cased-sentiments-student",
    top_k=None,
)
```

```python
with open("../goldenStandard/Dagmara/Annotation_sample_1.txt", "r") as f:
    sample_1 = [l.replace("Name: ", "", 1).replace("\u200b", "").strip() for l in f.readlines() if "Name:" in l]
print(len(sample_1), sample_1)
```

```python
with open("../goldenStandard/Dagmara/Annotation_sample_2.txt", "r") as f:
    sample_2 = [l.replace("Name: ", "", 1).replace("\u200b", "").strip() for l in f.readlines() if "Name:" in l]
print(len(sample_2), sample_2)
```

```python
with open("../goldenStandard/Dagmara/english_top_hits_2010s_moods.txt", "r") as f:
    sample_3 = [l.replace('"name": ', "", 1).replace('"', '').replace(',', '').strip() for l in f.readlines() if '"name": ' in l]
print(len(sample_3), sample_3)
```

```python
# Titles of songs to annotate 
samples = sample_1 + sample_2 + sample_3
print(len(samples), samples)
```

```python
with open("../Genius/english_song_lyrics/top_hits_2000__2009.json.json", "r") as f:
    songs_1 = json.load(f)["item"]
print(len(songs_1), songs_1)
```

```python
with open("../Genius/english_song_lyrics/english_top_hits_2010s.json.json", "r") as f:
    songs_2 = json.load(f)["item"]
print(len(songs_2), songs_2)
```

```python
with open("../Genius/english_song_lyrics/english_top-hits_2022-23.json.json", "r") as f:
    songs_3 = json.load(f)["item"]
print(len(songs_3), songs_3)
```

```python
# All songs
songs = songs_1 + songs_2 + songs_3
print(len(songs), songs)
```

```python
# Manual alteration
samples.remove("Chasing The Sun")
samples.append("Chasing the Sun")
```

```python
matches = [ s["name"] for s in songs if s["name"] in samples ]
missing = [s for s in samples if s not in matches]
print(missing)
```

```python
# Sentiment of translated song
def sentiment(lyrics: list[str]):
    l = " ".join(lyrics)[:1000]
    r = distilled_student_sentiment_classifier(l)[0]
    return sorted(r, key= lambda d: d["score"], reverse=True)[0]["label"]
```

```python
ratings = []
for s in songs:
    if s["name"] in matches:
        ratings.append(
            {
                "name":               s["name"],
                "artist":             s["artist"],
                "mood":     sentiment(s["lyrics"]),
                "lyrics":             s["lyrics"],
            }
        )
```
